## Weather dataset

```python
weather.info()
```

```python
weather.sample(20)
```

While info shows there are 20517 all non nulls, when displayed a sample 20 entries table we can notice there are missing values which need to be imputed.

In addition codesum feature is string and in order to process this dataset in a better way it need to be replaced into int.

```python
# explore the unique values in each column in order to notice out all the values which we need to impute
for col in weather.columns:
    print(col, weather[col].unique())
    print("----------------------------------------")
```

We noticed this values which we have to impute:

'M', ' ', '-'will be imputed into  null (Arbitary value imputation).

'T', ' T' stands for Trace which is a very small positive number and therefore will be imputed into 0.001 (Arbitary value imputation).

codesum will be used to generate a new numeric feature WeatherCode for unique codes.

```python
def clean_data(df):
    df['date'] = pd.to_datetime(df['date'])
    WeatherCode = ['FC', 'TS', 'GR', 'RA', 'DZ', 'SN', 'SG', 'GS', 'PL', 'IC', 'FG', 'BR', 'UP', 'HZ', 'FU', 'VA', 
               'DU', 'DS', 'PO', 'SA', 'SS', 'PY', 'SQ', 'DR', 'SH', 'FZ', 'MI', 'PR', 'BC', 'BL', 'VC', '+']
    df['WeatherCode'] = pd.Series(list(map(lambda x: sum(list(map(lambda y: x.find(y) >= 0, WeatherCode))), df.codesum)), index = df.index)
    for col in df.columns:
        df = df.replace(('-', ' '), np.nan)
        df = df.replace('M', np.nan)
        df = df.replace(('T', '  T'), 0.001)
        
        
    num_attr = ['tmax', 'tmin', 'tavg', 'dewpoint', 'wetbulb', 'heat', 'cool','sunset', 'sunrise', 'snowfall', 
                'preciptotal', 'stnpressure','sealevel', 'resultspeed', 'resultdir', 'avgspeed']
    df[num_attr] = df[num_attr].astype(float)
    df = df.loc[:,df.isnull().sum()< df.shape[0]*.5]
    return df


weather = clean_data(weather)
```

Now we will check again for null values

```python
weather.isna().sum()
```

We have null values and will have to impute them.

## Weather dataset

1- Drop features which have less than 65% non null

2- Drop entries which have less than 65% non null

3- impute the rest of null values using KNN Imputer in order to keep the data as close as possible to the raw data

```python
# 1-
weather.dropna(axis=1, thresh= len(weather)*0.65, how='any', subset=None, inplace=True)
weather = weather.reset_index(drop=True)

# 2-
weather.dropna(axis=0, thresh= len(weather.columns)*0.65, how='any', subset=None, inplace=True)
weather = weather.reset_index(drop=True)

display(weather.info())
```

```python
display(weather.sample(20))
```

```python
knnimputer = KNNImputer(n_neighbors=7)
noNullWeather = pd.DataFrame(knnimputer.fit_transform(weather.drop(columns=['date'])))
noNullWeather.set_axis(['station_nbr', 'tavg', 'cool', 'preciptotal', 'stnpressure', 'sealevel', 'resultdir', 'avgspeed', 'WeatherCode', 'target'], axis=1, inplace=True)
noNullWeather['date'] = weather['date']
display(noNullWeather.sample(20))
```

```python
noNullWeather.isna().sum()
```

Weather dataset cleaned, after features selection and imputations.

# Section C (Unit sales Prediction)

```python
salesRain = sales.merge(key)
salesRain = salesRain.merge(noNullWeather[['date','station_nbr','target']])
display(salesRain.sample(20))
```

### Creating train,test,validate sets

```python
key_sum = salesRain[salesRain['item_nbr'].isin([5,6,9,16,45])]
restItems = salesRain[~salesRain['item_nbr'].isin([5,6,9,16,45])]
X_test = key_sum.drop(columns=['units','date'])
y_test = key_sum['units']
X = restItems.drop(columns=['units','date'])
y = restItems['units']
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.33, random_state=42)
```

## Decision Tree Regressor

```python
# define model
dtr = DecisionTreeRegressor(random_state=42)

# define parameter grid
parameters_grid = {
    'max_depth': [7, 9, 11],
    'min_samples_split': [5, 10, 20],
    'min_samples_leaf': [2, 4, 6],
}

# define grid search
dtr_grid_search = GridSearchCV(estimator=dtr, param_grid=parameters_grid, cv=10)

# fit estimator
dtr_grid_search.fit(X_train, y_train)

# get best estimator
dtr_best = dtr_grid_search.best_estimator_

# print best parameters
pd.DataFrame.from_dict(dtr_grid_search.best_params_, orient='index', columns=['Selected Value']).T

# predict using model
dtrPreds = dtr_best.predict(X_test)

# print model's evaluation metrics
print("Decision Tree Regressor Mean_squared_error: ", mean_squared_error(y_test, dtrPreds))
print("Decision Tree Regressor R2_score: ", r2_score(y_test, dtrPreds))
```

## Adaboost Regressor

```python
# define model
adr = AdaBoostRegressor(random_state=42)

# define parameter grid
parameters_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.1, 0.2, 0.3]
}

# define grid search
adr_grid_search = GridSearchCV(estimator=adr, param_grid=parameters_grid, cv=10)

# fit estimator
adr_grid_search.fit(X_train, y_train)

# get best estimator
adr_best = adr_grid_search.best_estimator_

# print best parameters
pd.DataFrame.from_dict(adr_grid_search.best_params_, orient='index', columns=['Selected Value']).T

# predict using model
adrPreds = adr_best.predict(X_test)

# print model's evaluation metrics
print("Decision Tree Regressor Mean_squared_error: ", mean_squared_error(y_test, adrPreds))
print("Decision Tree Regressor R2_score: ", r2_score(y_test, adrPreds))
```

# Section D (Rainy Day Prediction)

### Creating train and test sets

```python
X = salesRain.drop(columns=['target','date'])
y = salesRain['target']
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

## Decision Tree Classifier

```python
# define model
dtc = DecisionTreeClassifier(random_state=42)

# define parameter grid
parameters_grid = {
    'max_depth': [7, 9, 11],
    'min_samples_split': [5, 10, 20],
    'min_samples_leaf': [2, 4, 6],
}

# define grid search
dtc_grid_search = GridSearchCV(estimator=dtc, param_grid=parameters_grid, cv=10)

# fit estimator
dtc_grid_search.fit(X_train, y_train)

# get best estimator
dtc_best = dtc_grid_search.best_estimator_

# print best parameters
pd.DataFrame.from_dict(dtc_grid_search.best_params_, orient='index', columns=['Selected Value']).T

# predict using model
dtcPreds = dtc_best.predict(X_test)

# print model's evaluation metrics
tn, fp, fn, tp = confusion_matrix(y_test, dtcPreds).ravel()
sensitivty = tp/(tp+fn)
specificity = tn/(tn+fp)
cmatrix = confusion_matrix(y_test, dtcPreds)
print(cmatrix)
print("Decision Tree Classifier Accuracy: ", accuracy_score(y_test, dtcPreds))
print("Decision Tree Classifier Sensitivity: ", sensitivty)
print("Decision Tree Classifier Specificity: ", specificity)
```
