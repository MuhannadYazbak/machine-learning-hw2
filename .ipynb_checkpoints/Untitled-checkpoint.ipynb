# Weather dataset (Visualization + preprocess):

```python
weather = pd.read_csv("weather.csv")
print("Weather dataset info:")
display(weather.info())
print("Weather dataset preview:")
display(weather)
```

All the columns except for station_nbr are of type object so we need to explore the unique values and cast to suitable type and deal with any missing value

```python
for col in weather.columns:
    print(col, weather[col].unique())
```

```python
# Cleaning method to replace all the bad values noticed above ('T', ' T', ' ', '-', 'M') and replace each unique codesum with an int (WeatherCode)
def clean_data(df):
    df['date'] = pd.to_datetime(df['date'])
    df['dateAsNumeric'] = pd.to_numeric(df['date'])
    WeatherCode = ['FC', 'TS', 'GR', 'RA', 'DZ', 'SN', 'SG', 'GS', 'PL', 'IC', 'FG', 'BR', 'UP', 'HZ', 'FU', 'VA', 
               'DU', 'DS', 'PO', 'SA', 'SS', 'PY', 'SQ', 'DR', 'SH', 'FZ', 'MI', 'PR', 'BC', 'BL', 'VC', '+']
    df['WeatherCode'] = pd.Series(list(map(lambda x: sum(list(map(lambda y: x.find(y) >= 0, WeatherCode))), df.codesum)), index = df.index)
    for col in df.columns:
        df = df.replace(('-', ' '), np.nan)
        df = df.replace('M', np.nan)
        df = df.replace(('T', '  T'), 0.001)
        
        
    num_attr = ['tmax', 'tmin', 'tavg', 'dewpoint', 'wetbulb', 'heat', 'cool','sunset', 'sunrise', 'snowfall', 
                'preciptotal', 'stnpressure','sealevel', 'resultspeed', 'resultdir', 'avgspeed']
    df[num_attr] = df[num_attr].astype(float)
    df = df.loc[:,df.isnull().sum()< df.shape[0]*.5]
    return df


weather = clean_data(weather)
display(weather.info())
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(weather.corr(), annot=True)
plt.title("Weather dataset features correlation")
plt.show()
```

### Weather dataset preprocessing steps :
1- we drop highly correlated features leaving only one member of each high correlation group

2- we drop features (columns) that does not have 65% or more non-null values

3- we drop entries (rows) that does not have 65% or more non-null features

4- the rest of null values will be imputed by KNNImputer in order to pick the closet prediction

```python
# 1
weather = weather.drop(columns=["tmax","tmin","heat","wetbulb","dewpoint", "resultspeed"])
# 2
weather.dropna(axis=1, how='any', thresh=len(weather)*0.65, subset=None, inplace=True)
weather = weather.reset_index(drop=True)
# 3
weather.dropna(axis=0, how='any', thresh=len(weather.columns)*0.65, subset=None, inplace=True)
weather = weather.reset_index(drop=True)
display(weather.info())
```
