# Section B (Preprocessing)

## Weather dataset

```python
# we start with date as numeric value
weather['dateAsNumber'] = pd.to_datetime(weather['date'])
weather['dateAsNumber'] = pd.to_numeric(weather['dateAsNumber'])

# we iterate each column to check what unique values it have and decide on suitable imputation technique 
for col in weather.columns:
    print(col, weather[col].unique())
```

```python
# Cleaning method to replace all the bad values noticed above ('T', ' T', ' ', '-', 'M') and replace each unique codesum with an int
def clean_data(df):
    WeatherCode = ['FC', 'TS', 'GR', 'RA', 'DZ', 'SN', 'SG', 'GS', 'PL', 'IC', 'FG', 'BR', 'UP', 'HZ', 'FU', 'VA', 
               'DU', 'DS', 'PO', 'SA', 'SS', 'PY', 'SQ', 'DR', 'SH', 'FZ', 'MI', 'PR', 'BC', 'BL', 'VC', '+']
    df['WeatherCode'] = pd.Series(list(map(lambda x: sum(list(map(lambda y: x.find(y) >= 0, WeatherCode))), df.codesum)), index = df.index)
    for col in df.columns:
        df = df.replace(('-', ' '), np.nan)
        df = df.replace('M', np.nan)
        df = df.replace(('T', '  T'), 0.001)
        
        
    num_attr = ['tmax', 'tmin', 'tavg', 'dewpoint', 'wetbulb', 'heat', 'cool','sunset', 'sunrise', 'snowfall', 
                'preciptotal', 'stnpressure','sealevel', 'resultspeed', 'resultdir', 'avgspeed']
    df[num_attr] = df[num_attr].astype(float)
    df = df.loc[:,df.isnull().sum()< df.shape[0]*.5]
    return df


clean_weather = clean_data(weather)
display(clean_weather.info())
```

```python
# now we want to drop any entry (row) which has less than 11 non-null values (Features)
clean_weather.dropna(axis=0, how='any', thresh=11, subset=None, inplace=True)
clean_weather = clean_weather.reset_index(drop=True) # reset the index to eliminate gaps
display(clean_weather.info())
```

```python
# now we do similair drop but this time for features (columns) which has less 60% non-null values (entries)
clean_weather.dropna(axis=1, how='any', thresh=19605*0.6, subset=None, inplace=True)
display(clean_weather.info())
```

```python
# the rest of null values, will be imputated by Statistical estimator (mean)
for col in clean_weather.columns:
    if  not 'date':
        clean_weather = clean_weather.replace(np.nan, clean_weather[col].mean())

    
display(clean_weather.info())
```

We replaced unique values with null execpt for 'T' which was replaced with a small positive float.

Next now eliminated all null values, we dropped rows with less than 11 non-null features, also dropped columns with less than 60% non-null entries. as these null values if we imputed it will distort the distrubition.

As a last step we imputed the rest null values which survived the drops by their column mean.

The other data sets needs no preprocessing (only date column as numeric).

Now, we can visualize the weather dataset in a better way

```python
plt.figure(figsize=(15,10))
sns.heatmap(clean_weather.corr(), annot=True)
plt.title("Features correlation in Weather")
plt.show()
```

As we can notice, the features (tmax,tmin,tavg,dewpoint,wetbulb and heat) have very strong correlations as well as a very strong correlation between the features (resultspeed and avgspeed). We can choose one of each group to keep and drop the other features as they will give the same effect.

# Section C (Unit sales predictions) :

```python
'''
1- we will filter  the products, keeping only products of interest hence (5,6,9,16,45).
2- we will group by store_nbr in order to tell in each store how many units of the interest products were sold.
3- we will group further more, this time by date, thus now we can tell in a given date and store how many units were sold.
4- we will add a new row with the item_nbr as -99 (key_sum )
'''
items = [5,6,9,16,45]
itemsDF = sales[sales["item_nbr"].isin(items)]
itemsDF = itemsDF.groupby(['store_nbr', 'date'])
for grp in itemsDF.groups:
    itemsDateDF = itemsDF.agg({'units': 'sum'})

itemsDateDF.to_csv('sectionC.csv', index=False)  
        
```
