## Weather dataset

```python
display(weather.info())
```

As can be noticed all coulmns but station_nbr are of type object/string. we need to further explore them for any special values
that should be replaced by null before any further preprocessing.

```python
# We start with date by adding numericDate Column
weather['numericDate'] = pd.to_numeric(pd.to_datetime(weather['date']))
for col in weather.columns:
    print(col)
    print(weather[col].unique())
    print("--------------------------------------")
```

As expected, there are  values that need to be replaced ('M', ' ', 'T', ' T', '-').

Instructions say that 'T' stands for trace which is a very small positive number (0.001), and says nothing else regarding the other special values noticed which therfore will be replaced by null.

In addition, codesum column will be dropped and replaced with a numeric column identifying the weather ('WeatherCode').

```python
def clean_data(df):
    WeatherCode = ['FC', 'TS', 'GR', 'RA', 'DZ', 'SN', 'SG', 'GS', 'PL', 'IC', 'FG', 'BR', 'UP', 'HZ', 'FU', 'VA', 
               'DU', 'DS', 'PO', 'SA', 'SS', 'PY', 'SQ', 'DR', 'SH', 'FZ', 'MI', 'PR', 'BC', 'BL', 'VC', '+']
    df['WeatherCode'] = pd.Series(list(map(lambda x: sum(list(map(lambda y: x.find(y) >= 0, WeatherCode))), df.codesum)), index = df.index)
    for col in df.columns:
        df = df.replace(('-', ' '), np.nan)
        df = df.replace('M', np.nan)
        df = df.replace(('T', '  T'), 0.001)
        
        
    num_attr = ['tmax', 'tmin', 'tavg', 'dewpoint', 'wetbulb', 'heat', 'cool','sunset', 'sunrise', 'snowfall', 
                'preciptotal', 'stnpressure','sealevel', 'resultspeed', 'resultdir', 'avgspeed']
    df[num_attr] = df[num_attr].astype(float)
    df = df.loc[:,df.isnull().sum()< df.shape[0]*.5]
    return df


weather = clean_data(weather)
display(weather.info())
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(weather.corr(), annot=True)
plt.title("Weather features correlations")
plt.show()
```

# Section B (preprocessing)

## Weather dataset

#### Weather preprocessing steps

1- Drop highly correlated features as they will not have any adding benefit.

2- Drop any feature (column) that have less than 65% non-nulls.

3- Drop any entry (row) that have less than 65% non-nulls.

4- The rest of null values will be imputed by KNNImputer in order to keep the data as close to raw data as possible.

Steps 2 and 3 helps to keep the data clean, as too many imputations in the same entry/feature may cause distortations. 

```python
# 1-
weather = weather.drop(columns=["tmax","tmin","heat","wetbulb","dewpoint", "resultspeed"])

# 2-
weather.dropna(axis=1, how='any', thresh=len(weather)*0.65, subset=None, inplace=True)
weather = weather.reset_index(drop=True) # reset the index to leave to no gaps

# 3-
weather.dropna(axis=0, how='any', thresh=len(weather.columns)*0.65, subset=None, inplace=True)
weather = weather.reset_index(drop=True) # reset the index to leave no gaps

display(weather.info())
display(weather.sample(10))
```

```python
# 4- 
weatherNoDate = weather.copy().drop(columns=['date'])
imputer = SimpleImputer(missing_values= np.nan, strategy='mean')
imputer.fit_transform(weatherNoDate)
```

```python
for col in weatherNoDate.columns:
    print(col, 'has a total of', weatherNoDate[col].isna().sum(), ' missing values')
```

Now, weather dataset is ready for more visualization as we dealed with the missing values, dropped features/entries with too many nulls, transformmed columns to numeric types and imputed the rest nulls.

# Section C (key_sum units prediction)

```python
# dataframe containing only items (5,6,916,45) which constructs key_sum and will be used as test set
key_sum = sales[sales['item_nbr'].isin([5,6,9,16,45])]
display(key_sum)
```

```python
# dataframe containing all of the rest items (not in key_sum) will be used as train/validation set
restItems = sales[~sales['item_nbr'].isin([5,6,9,16,45])]
display(restItems)
```

```python
print("unique item_nbr check\nKey_sum:")
display(key_sum['item_nbr'].unique())
print("restItems:")
display(restItems['item_nbr'].unique())
```

```python
X = restItems.drop(columns=['units','date','season'])
Y = restItems['units']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
display(X_test.sample(10))
```

```python
display(y_test.sample(10))
```

```python
X_key_sum = key_sum.drop(columns=['units','date','season'])
y_key_sum = key_sum['units']
dtRegressor = DecisionTreeRegressor(random_state= 50)
dtRegressor.fit(X_train, y_train)
preds = dtRegressor.predict(X_key_sum)
mse = mean_squared_error(y_key_sum, preds)
print(mse)
```
