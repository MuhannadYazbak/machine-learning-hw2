```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sales = pd.read_csv("sales.csv")
display(sales.head(10))
```

```python
display(sales.describe())
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(sales.corr(), annot=True)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=sales, x="store_nbr", y="item_nbr", hue="units")
plt.grid()
plt.title("store_nbr/item_nbr hue=units", fontsize=20, color="green")
plt.show()
```

```python
weather = pd.read_csv("weather.csv")
display(weather.head(10))
```

```python
display(weather.describe())
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(weather.corr(), annot=True)
plt.show()
```

```python
prec = weather["preciptotal"]
prec[10:50]
```

```python
weather['preciptotal'] = np.where(weather.preciptotal == '  T', '0.00', weather.preciptotal)
weather['preciptotal'] = np.where(weather.preciptotal == 'M', '20.00', weather.preciptotal)
weather['preciptotal'] = pd.to_numeric(weather['preciptotal'])
weather['preciptotal'][0:50]
```

```python
display(weather.describe())
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(weather.corr(), annot=True)
plt.show()
```

```python
weather['date'] = pd.to_datetime(weather['date'])
```

```python
display(weather.describe())
display(weather['date'][0:50])
#sns.scatterplot(weather, x="date", y="station_nbr", hue="preciptotal")
#plt.show()
```

```python
weather['tmax'] = np.where(weather.tmax == 'M', 100, weather.tmax)
weather['tmax'] = pd.to_numeric(weather['tmax'])
```

```python
display(weather['tmax'][0:100])
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(weather.corr(), annot=True)
plt.show()
```

```python
weather['tmin'] = np.where(weather.tmin == 'M', 10, weather.tmin)
weather['tmin'] = pd.to_numeric(weather['tmin'])
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(weather.corr(), annot=True)
plt.show()
```
